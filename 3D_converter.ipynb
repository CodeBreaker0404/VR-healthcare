```python


# Convert NumPy array to VTK image
data = data.astype(np.float32)  # Make sure it's float32
flat_data = data.ravel(order='F')  # Flatten in Fortran order
vtk_data = numpy_support.numpy_to_vtk(num_array=flat_data, deep=True, array_type=vtk.VTK_FLOAT)

# Create VTK image data
vtk_img = vtk.vtkImageData()
vtk_img.SetDimensions(data.shape)
vtk_img.SetSpacing(img.header.get_zooms())
vtk_img.GetPointData().SetScalars(vtk_data)

# Volume mapper
volume_mapper = vtk.vtkSmartVolumeMapper()
volume_mapper.SetInputData(vtk_img)

# Opacity transfer function
opacity_func = vtk.vtkPiecewiseFunction()
opacity_func.AddPoint(np.min(data), 0.0)
opacity_func.AddPoint(np.percentile(data, 50), 0.1)
opacity_func.AddPoint(np.max(data), 0.8)

# Color transfer function
color_func = vtk.vtkColorTransferFunction()
color_func.AddRGBPoint(np.min(data), 0.0, 0.0, 0.0)
color_func.AddRGBPoint(np.percentile(data, 50), 0.5, 0.5, 0.8)
color_func.AddRGBPoint(np.max(data), 1.0, 1.0, 1.0)

# Volume property
volume_prop = vtk.vtkVolumeProperty()
volume_prop.SetColor(color_func)
volume_prop.SetScalarOpacity(opacity_func)
volume_prop.ShadeOn()
volume_prop.SetInterpolationTypeToLinear()

# Create volume
volume = vtk.vtkVolume()
volume.SetMapper(volume_mapper)
volume.SetProperty(volume_prop)

# Renderer and window
renderer = vtk.vtkRenderer()
renderer.AddVolume(volume)
renderer.SetBackground(0.1, 0.1, 0.1)

render_window = vtk.vtkRenderWindow()
render_window.AddRenderer(renderer)
render_window.SetSize(800, 800)

interactor = vtk.vtkRenderWindowInteractor()
interactor.SetRenderWindow(render_window)

# Start interaction
render_window.Render()
interactor.Start()
```

```python
# Import necessary Slicer modules (in Slicer's Python environment, 'slicer' and 'vtk' are available)
import slicer, vtk
import os

# File paths – update these paths for your data
inputVolumePath = r"C:/path/to/inputVolume.nii"   # Path to the input NIfTI file
outputModelPath = r"C:/path/to/outputModel.obj"   # Desired path for the output OBJ file

# Validate input file exists
if not os.path.exists(inputVolumePath):
    raise FileNotFoundError(f"Input volume not found: {inputVolumePath}")
# Ensure the output directory exists
outputDir = os.path.dirname(outputModelPath)
if outputDir and not os.path.isdir(outputDir):
    os.makedirs(outputDir)

# 1. Load the NIfTI volume into a Slicer volume node
volumeNode = slicer.util.loadVolume(inputVolumePath, returnNode=True)[1]
if volumeNode is None:
    raise RuntimeError(f"Failed to load volume from {inputVolumePath}")
print(f"Volume loaded: {volumeNode.GetName()}")

# 2. Create a new segmentation node and add an empty segment
segmentationNode = slicer.mrmlScene.AddNewNodeByClass("vtkMRMLSegmentationNode", "Segmentation")
segmentationNode.CreateDefaultDisplayNodes()  # Prepare display properties (so the segment will be visible)
# Set the segmentation's reference geometry to match the volume
segmentationNode.SetReferenceImageGeometryParameterFromVolumeNode(volumeNode)
# Add a segment to hold the thresholded structure
segmentId = segmentationNode.GetSegmentation().AddEmptySegment("ThresholdSegment")
print(f"Created segmentation with segment ID: {segmentId}")

# 3. Set up the Segment Editor to apply thresholding
segmentEditorWidget = slicer.qMRMLSegmentEditorWidget()
segmentEditorWidget.setMRMLScene(slicer.mrmlScene)
segmentEditorNode = slicer.mrmlScene.AddNewNodeByClass("vtkMRMLSegmentEditorNode")
segmentEditorWidget.setMRMLSegmentEditorNode(segmentEditorNode)
segmentEditorWidget.setSegmentationNode(segmentationNode)
segmentEditorWidget.setMasterVolumeNode(volumeNode)

# Choose a basic intensity range for thresholding.
# (Adjust these values based on the volume's intensity to extract the desired structure.)
lowerThreshold = 100   # example minimum intensity value
upperThreshold = 1000  # example maximum intensity value

# Activate the Threshold effect and set its parameters
segmentEditorWidget.setActiveEffectByName("Threshold")
effect = segmentEditorWidget.activeEffect()
effect.setParameter("MinimumThreshold", str(lowerThreshold))
effect.setParameter("MaximumThreshold", str(upperThreshold))
effect.self().onApply()  # Apply the threshold - this fills the segment with the thresholded region
print(f"Applied threshold: kept intensities between {lowerThreshold} and {upperThreshold}")

# Remove the Segment Editor nodes (no longer needed after applying the effect)
segmentEditorWidget.setActiveEffect(None)
segmentEditorWidget = None
slicer.mrmlScene.RemoveNode(segmentEditorNode)

# 4. Generate a 3D surface (closed surface representation) from the segmented labelmap
segmentationNode.CreateClosedSurfaceRepresentation()  
# (The closed surface is needed to get a 3D mesh for export. This step ensures the segment has a surface mesh.)

# 5. Export the segmentation's 3D surface to an OBJ file
# Get the VTK polydata of the segment’s surface
surfaceMesh = segmentationNode.GetClosedSurfaceRepresentation(segmentId)
if surfaceMesh is None:
    raise RuntimeError("No surface mesh found - the segmentation might be empty or not converted to closed surface.")

# Use vtkOBJWriter to write the polydata to an .obj file
objWriter = vtk.vtkOBJWriter()
objWriter.SetInputData(surfaceMesh)
objWriter.SetFileName(outputModelPath)
if not objWriter.Write():
    raise RuntimeError(f"Failed to write OBJ file to {outputModelPath}")

print(f"3D model successfully saved as: {outputModelPath}")
```
